```python
import sys
import pandas as pd
import os

sys.path.append('..')

import donedeal as dd

```

```python
scraper = dd.CarScraper()
scraper.set_make_model('audi', 'a4')
scraper.set_year(2016,2017)
scraper.scrape()
```

```python
scraper.DataFrame[dd.constants.RECOMMENDED_COLUMNS]
```

```python
import requests
import re

from donedeal.constants import DOMAIN, REQUEST_HEADERS, CAR_HREF_PATTERN 
from donedeal import utils
from pandas import DataFrame

self = scraper
batch_size = 100


pageinated_search_urls:list[str] = utils.get_pageinated_urls(self.url)
self.hrefs:list[str] = [DOMAIN+href for href in utils.get_car_hrefs(urls=pageinated_search_urls)]
print("retrieved {0} urls matching your criteria".format(self.hrefs.__len__()))
self.json:list[dict] = utils.get_json_data_from_ad_urls(self.hrefs,batch_size=batch_size)
print("retrieved {0} jsons".format(self.json.__len__()))
self.DataFrame:DataFrame = utils.json_to_df(self.json)
print("converted jsons to DataFrame")
```
